```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd 
```

```python
# a) Creación de vectores:
a = np.array([3.1, 1, -0.5, -3.2, 6])
b = np.array([1, 3, 2.2, 5.1, 1])
print("a) Vectores creados:")
print("a =", a)
print("b =", b)

# b) Multiplicación escalar (producto punto):
m_e = np.dot(a, b)
print("\nb) Producto punto a·b =", m_e)
print("Para multiplicar dos vectores con producto punto, deben tener la misma dimensión.")

# c) Multiplicación punto a punto (elemento a elemento):
m_p = a * b
print("\nc) Producto elemento a elemento a*b =", m_p)

# d) Construcción de matriz A:
A = np.array([
    [2, -1, -3],
    [4, 1.5, -2.5],
    [7.3, -0.9, 0.2]
])
print("\nd) Matriz A:")
print(A)

# e) Transpuesta de A
A_transpuesta = A.T
print("\ne) Transpuesta de A:")
print(A_transpuesta)

# f) Funciones de NumPy:
print("\nf) Ejemplos de funciones NumPy:")
print("ones(): Crea un arreglo de unos, segun la dimensión que se le indique.\n", np.ones(A.shape))
print("\nround(): Redondea los elementos del array al valor mas cercano y la cantidad de decimales.\n", np.round(A, 0))
print("\nceil(): Redondea hacia arriba al entero mas cercano.\n", np.ceil(A))
print("\nfloor(): Redondea hacia abajo al entero mas cercano.\n", np.floor(A))

# g) Acceder a primera fila, tercera columna
valor = A[0, 2]
print("\ng) Valor primera fila, tercera columna:\n", valor)

# h) Obtener segunda fila
fila2 = A[1, :]
print("\nh) Segunda fila de A:\n", fila2)

# i) Dimensiones de la matriz
dimensiones = A.shape
print("\ni) Dimensiones de A:\n", dimensiones)

# j) Función seno y[n]
n = np.arange(0, 101)
y = np.sin(np.pi * 0.12 * n)

# k) Función coseno y2[n]
y2 = np.cos(2 * np.pi * 0.03 * n)

# l) Señales combinadas
s = y + y2
t = y * y2

# m) Gráfica de y[n] y y2[n]
plt.figure(figsize=(10, 4))
plt.plot(n, y, 'b-', label='y[n] = sin(π*0.12n)')
plt.plot(n, y2, 'r--', label='y2[n] = cos(2π*0.03n)')
plt.title('Señales')
plt.xlabel('Intervalo (n)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()

# n) Gráfica de s[n] y t[n]
plt.figure(figsize=(10, 4))
plt.plot(n, s, 'g-', label='s[n] = y[n] + y2[n]')
plt.plot(n, t, 'm--', label='t[n] = y[n] * y2[n]')
plt.title('señales')
plt.xlabel('Intervalo (n)')
plt.ylabel('Amplitud')
plt.legend()
plt.grid(True)
plt.show()
```

```python
'''PANDAS'''

# Punto 1)
def notas(dic_notas):
    serie_notas = pd.Series(dic_notas)
    resumen = pd.Series({
        'Mínima': serie_notas.min(),
        'Máxima': serie_notas.max(),
        'Media': serie_notas.mean(),
        'Desviación Típica': serie_notas.std()
    })
    return resumen

notas_curso = {
    'Juan': 7.5,
    'María': 8.2,
    'Pedro': 6.8,
    'Ana': 9.1,
    'Luis': 5.6
}

print("Resumen de notas:")
print(notas(notas_curso))
```

```python
#Punto 2)

# a) Cargar los datos.
df = pd.read_csv('datos.csv', sep=";")

# b) 5 primeras y últimas filas.
print("\nb) Primeras filas:\n",df.head())
print("\nUltimas filas:\n",df.tail())


# c) Borrar la columna "Unnamed: 0"
print("\nc) titulos:",df.columns.tolist())
if 'Unnamed: 0' in df.columns:
    df.drop('Unnamed: 0', axis=1, inplace=True)
    print("\nDataFrame después de eliminar 'Unnamed: 0':\n",df.head())


# d) Calcular el IMC (BMI)
# Organizar orden de las columnas:
df.rename(columns={
    'Weight': 'Height',
    'Height': 'Weight'
}, inplace=True)
print("\nd) Cambio de orden en la columna de peso y altura:\n",df.head())

# Caculo de IMC
imc = df['Weight'] / ((df['Height'] / 100) ** 2)
print("\nValores de IMC\n",imc)

# e) Calculo del IMC como BMI
df['BMI'] = df['Weight'] / ((df['Height'] / 100) ** 2)
print("\ne) DataFrame con columna BMI añadida:")
print(df.head())

# f) Categorización según BMI
def categorizar_bmi(bmi):
    if bmi < 18.5:
        return 'Bajo peso'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Sobrepeso'
    else:
        return 'Obesidad'

df['Categoría BMI'] = df['BMI'].apply(categorizar_bmi)
print("\nf) DataFrame con categorización BMI:\n", df)
```
