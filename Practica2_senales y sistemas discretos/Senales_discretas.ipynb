```python
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
```

```python
'''Punto 1'''
# EscalÃ³n unitario.
def escalon(n0, n1, n2):
    n = np.arange(n1, n2 + 1)
    x = (n >= n0).astype(int)
    return x, n
```

```python
'''Punto 2'''
# Rampa
def rampa(n0, n1, n2, m=1):
    n = np.arange(n1, n2 + 1)
    x = np.maximum(0, m * (n - n0))
    return x, n
```

```python
'''Punto 3'''

'''Numeral (a)'''
n = np.arange(-5, 6)
# Secuencia x1(n) = 3Î´(n+1) + 5Î´(n+3) + 3Î´(n+2) + 3Î´(n) + Î´(n)
x1 = 3 * (n == -1).astype(int) + \
     5 * (n == -3).astype(int) + \
     3 * (n == -2).astype(int) + \
     3 * (n == 0).astype(int) + \
     1 * (n == 0).astype(int)

# Grafica:
print("\na) Grafica x1(n):")
plt.figure(figsize=(8, 6))
plt.stem(n, x1, 'r')
plt.title("Secuencia x1(n)")
plt.xlabel("n")
plt.ylabel("X1")
plt.grid()
```

```python
'''Numeral (b)'''
def escalon(n0, n):
    return (n >= n0).astype(int)

n = np.arange(-6, 7)
u_n = escalon(0, n)

# Multiplicaciones y adelantos
u1 = -2 * np.roll(u_n, -4)
u2 = 3 * np.roll(u_n, -2)
x2 = u1 + u2

# Grafica:
print("\nb) Grafica x2(n):")
plt.figure(figsize=(8, 4))
plt.plot(n, x2, linestyle='-', marker='o', color='r')
plt.title("x2(n)")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.axhline(0, color='blue', linewidth=0.5)
plt.axvline(0, color='blue', linewidth=0.5)
plt.grid()
```

```python
'''Numeral (c)'''
n = np.arange(0, 101) 
x3 = np.exp(0.01 * n) * np.sin(2 * np.pi * n)

# Grafica:
print("\nc) Grafica x3(n):")
plt.figure(figsize=(8, 4))
plt.plot(n, x3, linestyle='-', color='r')
plt.title("x3(n)")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.axhline(0, color='blue', linewidth=0.5)
plt.axvline(0, color='blue', linewidth=0.5)
plt.grid()
```

```python
'''Numeral (d)'''
def escalon(n0, n):
    return (n >= n0).astype(int)

def rampa(n0, n):
    return ((n - n0) * (n >= n0)).astype(int)

n = np.arange(-10, 11)
x4 = 2 * rampa(-3,n) - rampa(2,n) - 5 * escalon(3, n)

# Grafica:
print("\nd) Grafica x4(n):")
plt.figure(figsize=(8, 4))
plt.plot(n, x4, '-o', color="r")
plt.title("x4(n)")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.axhline(0, color='blue', linewidth=0.5)
plt.axvline(0, color='blue', linewidth=0.5)
plt.grid()

```

```python
'''Punto 4'''
'''Numeral (a)'''

def rampa(n0, n):
    return ((n - n0) * (n >= n0)).astype(int)

n = np.arange(0, 18)

x = np.piecewise(n, [
    (n >= 0) & (n <= 5),
    (n >= 6) & (n <= 11),
    (n >= 12) & (n <= 17)
], [
    lambda n: rampa(0, n),
    lambda n: rampa(5, n),
    lambda n: rampa(10, n)
])

dx = np.diff(x, prepend=0)
print("\na) derivada de x:\n",dx)

'''Numeral (b)'''
# Grafica:
print("\nb) Graficas:")
plt.subplot(2, 1, 1)
plt.plot(n, x, '-o')
plt.title("x[n]")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(n, dx, '-o', color='r')
plt.title("Derivada de x[n]")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.grid()
plt.tight_layout()

```

```python
'''Punto 5'''
'''Numeral (a)'''

# Dominio:
n = np.arange(0, 23)

# DefiniciÃ³n de x(n) por tramos:
x = np.piecewise(n, [
    (n >= 0) & (n <= 5),     # parte1: 0,1,2,3,4,5
    (n >= 6) & (n <= 10),    # parte2: 4,3,2,1,0
    (n >= 11) & (n <= 15),   # parte3: 1,2,3,4,5
    (n >= 16) & (n <= 19),   # parte4: 5,5,5
    (n >= 19) & (n <= 22)    # parte5: 10,10,10,10
], [
    lambda n: n,                        
    lambda n: 10 - n,                   
    lambda n: n - 10,                   
    lambda n: 5,                        
    lambda n: 10                        
])

x5 = np.zeros_like(x)

for i in range(len(n)):
    x_n = x[i]
    x_n_4 = x[i - 4] if i - 4 >= 0 else 0 
    x5[i] = ((2 * x_n_4) + x_n)


print("\na) Graficas: x(n) y x5(n)")
plt.subplot(2, 1, 1)
plt.plot(n, x, '-o')
plt.title("x[n]")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.grid()

plt.subplot(2, 1, 2)
plt.plot(n, x5, '-o', color='r')
plt.title("Derivada de x5[n]")
plt.xlabel("n")
plt.ylabel("Amplitud")
plt.grid()
plt.tight_layout()

'''Numeral (b)'''
```
